```python
data = pd.read_csv("E:/kc_house_data.csv")
data
```

```python
data.describe()
```

```python
data.plot(x='price',y='sqft_lot' , style='o')
# plt.xlabel('sqft_lot')
# plt.ylabel('price')
```

```python
new_data = data.loc[:,['price','sqft_lot']]
```

```python
new_data
```

```python
d = new_data
```

```python
d.head()
```

```python
d['price']
```

```python
x = pd.DataFrame(d['sqft_lot'])
```

```python
y=pd.DataFrame(d['price'])
```

```python
x.size
```

```python
y.size
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2,random_state=1)
```

```python
print(x_train.size)
print(x_test.size)
print(y_train.size)
print(y_test.size)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor =LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
print(regressor.intercept_)
```

```python
print(regressor.coef_)
```

```python
y_pred = regressor.predict(x_test)
y_pred = pd.DataFrame(y_pred , columns = ['predicted'])
y_pred.index = x_test.index
y_pred
```

```python
test = x_test.join(y_test)
test = test.join(y_pred)
test
```
